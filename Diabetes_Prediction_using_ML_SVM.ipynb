Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Data Collection and Analysis

```python
#Loading the diabetes dataset to a pandas DataFrame
diabetes_dataset = pd.read_csv(r'/content/drive/MyDrive/Datasets/diabetes.csv')
diabetes_dataset.head()
```

```python
#number of rows and columns in this dataset
diabetes_dataset.shape
```

```python
#getting the stastical measures of the data
diabetes_dataset.describe
```

```python
diabetes_dataset[ 'Outcome'].value_counts ()
```

```python
#0 --> Non- Diabetic
#1 --> Diabetic
```

```python
diabetes_dataset.groupby('Outcome'  ).mean()
```

```python
#separating the data and labels
x = diabetes_dataset.drop(columns='Outcome',axis=1)
y = diabetes_dataset['Outcome']
```

```python
print(x)
```

```python
print(y)
```

Data Standardization

```python
scaler = StandardScaler()
```

```python
scaler.fit(x)
```

```python
standardized_data = scaler.transform(x)
```

```python
print(standardized_data)
```

```python
x = standardized_data
y = diabetes_dataset['Outcome']
```

```python
print(x)
print(y)
```

Splite the data into Train and Test

```python
x_train,x_test,y_train,y_test = train_test_split( x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

Traning the Model

```python
classifier = svm.SVC(kernel='linear')
```

```python
#traning the svm classifier
classifier.fit(x_train,y_train)
```

Model Evaluation

Finding the accuracy score

```python
#accuracy score on the traning data
x_train_prediction = classifier.predict(x_train)
```

```python
traning_data_accuracy = accuracy_score(x_train_prediction,y_train)
```

```python
print('Accuracy score of the traning data :',traning_data_accuracy)
```

```python
#accuracy score on the test data
x_test_prediction = classifier.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction,y_test)
```

```python
print('Accuracy score of the test data :',test_data_accuracy)
```

Making a Predictive System

```python
input_data = (1,103,30,38,83,43.3,0.183,33)

#changing the input data  to numpy array
input_data_as_numpy_array =  np.asarray(input_data)

#reshape the  array as we are predicting for one instance
input_data_reshape = input_data_as_numpy_array.reshape(1,-1)

#standardized the input data
std_data = scaler.transform(input_data_reshape)
print(std_data)

prediction = classifier.predict(std_data)
print(prediction)

if (prediction[0] == 0):
  print('The person is not diabetic.')
else:
  print('The person is diabetic.')
```
